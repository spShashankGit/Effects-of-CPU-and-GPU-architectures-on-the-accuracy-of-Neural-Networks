```python
def loadDataSet(filename):
    dataset = pd.read_csv(filename,delimiter=",",header=None)
    dataset = dataset.transpose()
    
    avg = 0
    avgLossPerEpoch = []
    max_epoch_num = 150
    avgDeno = len(dataset)/max_epoch_num
    
    for j in range(len(dataset)):
        avg = avg + float(dataset[0][j])
        if(j%avgDeno==0):
            avgLossPerEpoch.append(avg)
            avg = 0
    return avgLossPerEpoch
```

```python
def printFreq2(data):
    counter = 0
    res = []
    
    for i, vali in enumerate(data):
        refNum = vali
        print(type(refNum))
        for j in range(i,len(data)):
            if (round(refNum,1) == round(data[j], 1)):
                counter += counter 
        res.append({round(refNum,1):counter})
        counter = 0
        
    return res
```

```python
def printFreqIsClose(data):
    counter = 0
    res = []
    
    for i, vali in enumerate(data):
        refNum = vali
        print(type(refNum))
        for j in range(i,len(data)):
            if math.isClose(refNum,1) == round(data[j], 1)):
                counter += counter 
        res.append({round(refNum,1):counter})
        counter = 0
        
    return res
```
