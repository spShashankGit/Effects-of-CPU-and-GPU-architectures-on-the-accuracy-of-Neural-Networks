```python
#!conda install -y pandas
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%load_ext autotime
```

```python
def loadDataSet(filename):
    dataset = pd.read_csv(filename,delimiter=",",header=None)
    dataset = dataset.transpose()
    
    avg = 0
    avgLossPerEpoch = []
    max_epoch_num = 150
    avgDeno = len(dataset)/max_epoch_num
    
    for j in range(len(dataset)):
        avg = avg + float(dataset[0][j])
        if(j%avgDeno==0):
            avgLossPerEpoch.append(avg)
            avg = 0
    return avgLossPerEpoch
```

## LR 0.01

```python
dataset0 = loadDataSet('loss_values_0.csv')
```

```python
plt.figure(figsize=(30, 12)).suptitle('Learning Rate = 0.01',fontsize=45)
plt.plot(dataset0)
plt.show()
```
