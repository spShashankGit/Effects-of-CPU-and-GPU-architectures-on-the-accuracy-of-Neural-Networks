# Logistic Map

### $X_{n+1} = r X_{n}(1-X_{n})$

where $r$ is the rate



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
x_init = 0.4
iteration = 0
max_iteration = 100
r = 3.5
score = pd.DataFrame(columns =['Iter','Xn', 'r', 'Xn+1'],dtype=float)
scrn = np.array([])
scrnp1 = np.array([])
k = True
```

```python
def logisticMapCalc(x):
    global score
    global iteration
    global scrn
    global scrnp1
    global k
    
    xNew = r*x*(1-x)
    iteration += 1
    #print(iteration, " Xn = ", xNew)
    
    rowData = pd.DataFrame([[iteration,x,r,xNew]], columns=['Iter','Xn', 'r', 'Xn+1'],dtype=float)
    score = score.append(rowData,ignore_index=True)
    
    if(xNew == x and k):
        print("Convergence point is at epoch", iteration, "and the value is",x)
        k=False
        
    if (iteration <= max_iteration):
        logisticMapCalc(xNew)
    
```

```python
logisticMapCalc(x_init)
```

```python
fig, ax = plt.subplots(figsize=(10,10))
sns.lineplot(x="Iter",y="Xn+1", markers=True,data=score[0:25],err_style="bars")
sns.lineplot(x="Iter",y="Xn+1", markers=True,data=score[25:50],err_style="bars")
sns.lineplot(x="Iter",y="Xn+1", markers=True,data=score[50:75],err_style="bars")
sns.lineplot(x="Iter",y="Xn+1", markers=True,data=score[75:99],err_style="bars")
```
